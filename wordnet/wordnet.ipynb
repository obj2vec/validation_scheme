```python
from scipy.stats import spearmanr

f=open("SimLex-999.txt", 'r').readlines()

simlex_pairs=[]
wn_pairs=[]
for i in f[1:]:
    ii=i.split('\t')
    first_word=ii[0]
    second_word=ii[1]
    pos=ii[2]
    if pos == 'N':
        simlex_pairs.append(float(ii[3]))
        wn1 = wn.synset(first_word+'.n.01')
        wn2 = wn.synset(second_word+'.n.01')
        wn_pairs.append(wn1.path_similarity(wn2))
spearmanr(simlex_pairs, wn_pairs)
```

```python
from scipy.stats import spearmanr

f=open("SimLex-999.txt", 'r').readlines()

simlex_pairs=[]
wn_pairs=[]
for i in f[1:]:
    ii=i.split('\t')
    first_word=ii[0]
    second_word=ii[1]
    pos=ii[2]
    try:
        wn1 = wn.synset(first_word+'.'+pos.lower()+'.01')
        wn2 = wn.synset(second_word+'.'+pos.lower()+'.01')
        res=wn1.path_similarity(wn2)
        if res:
            wn_pairs.append(res)
            simlex_pairs.append(float(ii[3]))
    except:
        print(first_word+'.'+pos.lower()+'.01')
spearmanr(simlex_pairs, wn_pairs)
```
